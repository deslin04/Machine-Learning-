## checking for missing values

```python
data.isnull().sum()
```

```python
obcols=(data.dtypes[data.dtypes=='object']).index
```

```python
for x in obcols:
  print(x,(data[x]=='?').sum())
#checking for ? 

# 3 columns have ?
```

```python
#replace ? with mode of that column
workclass_mode = data['workclass'].mode()[0]
occupation_mode = data['occupation'].mode()[0]
native_country_mode = data['native-country'].mode()[0]
```

```python
#replacing ? for workclass with mode
data['workclass'][data['workclass'] == '?'] = workclass_mode
```

```python
data['occupation'][data['occupation'] == '?'] = occupation_mode
```

```python
data['native-country'][data['native-country'] == '?'] = native_country_mode
```

```python
#confrim missing value handing
for x in obcols:
  print(x,(data[x]=='?').sum())
```

```python
#To get integer columns
data.dtypes[data.dtypes=='int64'].index
```

```python
data['income'] = data['income'].map({'<=50K':0,'>50K':1})

#map function used to map each value into corresponing other value
```

# Simple Visualization
## matplotlib & seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Univariate Analysis

#### histogram

```python
#univariate analysis

#histogram for age

data['age'].hist()
plt.title("Histogram for Age",fontsize=20)
plt.xlabel("Age",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.yscale('linear')
plt.show()
```

```python
data['hours-per-week'].hist(figsize=(8,8),bins=[5,25,45,65,85,105])
plt.title("Histogram for Working Hours",fontsize=20)
plt.xlabel("Age",fontsize=16)
plt.ylabel("Count",fontsize=16)
plt.show()
```
